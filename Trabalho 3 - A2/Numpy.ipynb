```python
#2. Converta os dados relevantes em arrays NumPy (utilize o método .to_numpy())
df = pd.read_csv('Salary_Data.csv')

dadosArray = df.to_numpy()
print(dadosArray)
```

```python
#3. Calcule a média de salários.
df = pd.read_csv('Salary_Data.csv')

mediaSalary = df['Salary'].mean()
print(f"A média de salários é: {mediaSalary:.2f}")
```

```python
#4. Calcule a média de tempo de serviço.
df = pd.read_csv('Salary_Data.csv')

mediaTempo = df['YearsExperience'].mean()
print(f"A média de tempo de serviço é: {mediaTempo:.2f}")
```

```python
#5. Encontre o salário médio dos funcionários com mais de 5 anos de serviço.
df = pd.read_csv('Salary_Data.csv')

funcionariosMais5Anos = df[df['YearsExperience'] > 5]
mediaSalary = funcionariosMais5Anos['Salary'].mean()

print(f"O salário médio dos funcionários com mais de 5 anos de serviço é: {mediaSalary:.2f}")
```

```python
#6. Crie um gráfico de barras mostrando os salários referente ao tempo de serviço.
import matplotlib.pyplot as plt

df = pd.read_csv('Salary_Data.csv')
media = df.groupby('YearsExperience')['Salary'].mean()

plt.figure(figsize=(10, 6))
media.plot(kind='bar', color='skyblue')
plt.title('Salário Médio por Tempo de Serviço')
plt.xlabel('Tempo de Serviço (anos)')   
plt.ylabel('Salário Médio')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
7#. Considere um aumento de 10% para o salário de cada funcionário e apresente o resultado do aumento.
df = pd.read_csv('Salary_Data.csv')
df['Novo_Salario'] = df['Salary'] * 1.10

print(df[['YearsExperience', 'Salary', 'Novo_Salario']])

df.to_csv('Salary_Data_Aumentado.csv', index=False)
```
